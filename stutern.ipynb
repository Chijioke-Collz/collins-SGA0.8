```python
# Multivariate analysis of the different test scores (math, reading and writing) of both gender types using a barplot

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.barplot(x=sp.gender , y=sp["math_score"])
plt.title("Math scores")

plt.subplot(1,3,2)
sns.barplot(x=sp.gender , y=sp["reading_score"])
plt.title("Reading Scores")

plt.subplot(1,3,3)
sns.barplot(x=sp.gender , y=sp["writing_score"])
plt.title("Writing Scores")

plt.tight_layout()
plt.show()
```

```python
# Multivariate analysis of the different races or ethnicities or groups in relation to their performances in the test. This is done using barplots.

plt.figure(figsize=(13,4))

plt.subplot(1,3,1)
sns.barplot(x = sp['race/ethnicity'] , y= sp['reading_score'])
plt.title("Reading Scores")

plt.subplot(1,3,2)
sns.barplot(x = sp['race/ethnicity'] , y= sp['writing_score'])
plt.title("Writing Scores")

plt.subplot(1,3,3)
sns.barplot(x = sp['race/ethnicity'] , y= sp['math_score'])
plt.title("Math Scores")

plt.tight_layout()
plt.show()
```

```python
# Multivariate analysis of how different lunch packages affect student performances. This done using barcharts

plt.figure(figsize=(13,4))

plt.subplot(1,3,1)
sns.barplot(x = sp['lunch'] , y= sp['reading_score'], palette = 'viridis')
plt.title("Reading Scores")

plt.subplot(1,3,2)
sns.barplot(x = sp['lunch'] , y= sp['writing_score'], palette = 'viridis')
plt.title("Writing Scores")

plt.subplot(1,3,3)
sns.barplot(x = sp['lunch'] , y= sp['math_score'], palette = 'viridis')
plt.title("Math Scores")

plt.tight_layout()
plt.show()
```

```python
# Multivariate analysis of students based on the level of education of their parents (parentEdu).

plt.figure(figsize=(13,4))

plt.subplot(1,3,1)
sns.barplot(x = sp['parental_level_of_education'] , y= sp['reading_score'])
plt.xticks(rotation = 90)
plt.title("Reading Scores")

plt.subplot(1,3,2)
sns.barplot(x = sp['parental_level_of_education'] , y= sp['writing_score'])
plt.xticks(rotation=90)
plt.title("Writing Scores")

plt.subplot(1,3,3)
sns.barplot(x = sp['parental_level_of_education'] , y= sp['math_score'])
plt.xticks(rotation=90)
plt.title("Math Scores")

plt.tight_layout()
plt.show()
```

```python
# Performance of students based on their preparation for the test (testPreparation_course).

plt.figure(figsize=(13,4))

plt.subplot(1,3,1)
sns.barplot(x = sp['test_preparation_course'] , y= sp['reading_score'])
plt.title("Reading Scores")

plt.subplot(1,3,2)
sns.barplot(x = sp['test_preparation_course'] , y= sp['writing_score'])
plt.title("Writing Scores")

plt.subplot(1,3,3)
sns.barplot(x = sp['test_preparation_course'] , y= sp['math_score'])
plt.title("Math Scores")

plt.tight_layout()
plt.show()
```

```python
#To identify number of students that passed or failed the tests. Columns are created showing the grades of students for each test.
#Scores which are less than 40 (i.e <40) means that student has failed the test, otherwise the student has passed the test.

sp['math_grade'] = np.where(sp['math_score']<40,"fail" , "pass")
sp['reading_grade'] = np.where(sp['reading_score']<40, "fail" , "pass")
sp['writing_grade'] = np.where(sp['writing_score']<40, "fail" , "pass")
```

```python
# This is based on the level of education of their parents

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot(x = sp['parental_level_of_education'] ,hue= sp.math_grade ,palette="dark")
plt.xticks(rotation=90)
plt.title("Math Grade")

plt.subplot(1,3,2)
sns.countplot(x = sp['parental_level_of_education'] ,hue= sp.reading_grade ,palette="viridis")
plt.xticks(rotation=90)
plt.title("Reading Grade")

plt.subplot(1,3,3)
sns.countplot(x = sp['parental_level_of_education'] ,hue= sp.writing_grade ,palette="hot_r")
plt.xticks(rotation=90)
plt.title("Writing Grade")
plt.show()
```
